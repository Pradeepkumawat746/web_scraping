```python
import time 
from selenium import webdriver 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.select import Select
import pandas as pd 
```

```python
driver=webdriver.Chrome()
driver.maximize_window()



wait=WebDriverWait(driver,5)
def wait_for_page_to_load(driver, wait):
    page_title=driver.title
    try:
        wait.until(
            lambda d:d.execute_script("return document.readyState")=="complete"
      )
    except:
        print(f"the page{page_title} is not loaded")
    else:
        print(f"the page {page_title} is loaded")
    
url='https://www.scrapethissite.com/pages/forms/'
driver.get(url)  
wait_for_page_to_load(driver, wait)

button=driver.find_element(By.XPATH,'//*[@id="per_page"]')
dropdown=Select(button)
time.sleep(2)

dropdown.select_by_index(2)
time.sleep(2)

# web scraping 

data = []

while True:
    rows = driver.find_elements(By.CSS_SELECTOR, "table tbody tr")

    for row in rows:
        values = row.find_elements(By.TAG_NAME, "td")
        if len(values) < 9:
            continue
        stock = {
            "Team": values[0].text,
            "Year": values[1].text,
            "Wins": values[2].text,
            "Losses":values[3].text,
            "Win %":values[5].text,
            "GF": values[6].text,
            "GA": values[7].text
        }
        data.append(stock)

    # Try to click the 'next' button
    try:
        next_button = driver.find_element(By.LINK_TEXT, "»")
        if "disabled" in next_button.get_attribute("class"):
            break  # stop if disabled
        next_button.click()
        time.sleep(1)  # wait for page load
    except:
        break  # stop if button not found

driver.quit()


		
```
