```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://www.scrapethissite.com/pages/simple/'
```

```python
country=[]
capital=[]
population=[]
area=[]
```

```python
try:
    response=requests.get(url)
    response.raise_for_status()
except Exception as e:
    print(e)
else:
    response=requests.get(url)
    soup=BeautifulSoup(response.content,"html.parser")
    n=soup.find_all('h3')
    for i in n:
        con=i.get_text()
        country.append(con)
    
        capi=soup.find_all('div',class_='country-info')
    for i in capi:
        capitals=i.find('span',class_='country-capital').get_text(strip=True)
        populations=i.find('span',class_='country-population').get_text(strip=True)
        Areas =i.find('span',class_='country-area').get_text(strip=True)
        capital.append(capitals)
        population.append(populations)
        area.append(Areas)

clean=[item.replace('\n','').replace(',','')for item in country]
data={'country':clean,
      'capital':capital,
      'population':population,
      'Area':area
    }
data1=pd.DataFrame(data)
```
